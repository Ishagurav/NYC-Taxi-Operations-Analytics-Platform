# BRONZE LAYER — Raw Ingestion
# ================================================

```python
from pyspark.sql.functions import current_timestamp, lit
```

```python
bronze_df = spark.read.parquet(
    "/Volumes/nyc_taxi_catalog/bronze/bronze_raw/yellow_tripdata_2026-01.parquet"
)
```

```python
# Add audit columns
bronze_df = bronze_df \
    .withColumn("ingestion_timestamp", current_timestamp()) \
    .withColumn("source_file", lit("yellow_tripdata_2026-01.parquet"))
```

## # Save as Delta Table in Bronze Schema

```python

bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("nyc_taxi_catalog.bronze.bronze_taxi_data")

print(f"✅ Bronze Layer Done! Records: {bronze_df.count():,}")
```
